```python
# 필터링
# 성별이 남성인 데이터만 조회하기..

sample_1[sample_1["성별"]=="남성"] # Index를 기준으로 True/False 구분하여 필터링 실시

# iloc 인덱스 번호 / loc 인덱스 값 
# 인덱스 번호는 안보임 / 0, 2, 4 는 인덱스 값! 
```

```python
# 입국객수가 150000 이상인 데이터만 조회하기..
tf = sample_1["입국객수"] >= 150000
```

```python
sample_1[tf]
```

```python
# 성별이 여성이고, 입국객수가 150000 이상인 데이터 조회하기
sample_1[(sample_1["성별"] == "여성") & \
         (sample_1["입국객수"] >= 150000)]
```

```python
# 국적코드의 값이 A01 또는 A18인 데이터만 조회하기
sample_1[(sample_1["국적코드"] == "A01") | (sample_1["국적코드"] == "A18")]
```

```python
# isin()함수 : 같은 컬럼내에 여러개 값이 포함되어 있는지 확인할 때 사용(다른 컬럼일 때 "|" 사용)
sample_1[sample_1["국적코드"].isin(["A01", "A18"])]
```

```python
sample_1[(sample_1["국적코드"] != "A01") & (sample_1["국적코드"] != "A18")]
```

```python
sample_1[sample_1["국적코드"].isin(["A01", "A18"])==False]
```

```python
# sample_codemaster.xlsx 파일 읽어들인 후 조회하기
# 데이터프레임변수 : code_master

file_path = "./files/sample_codemaster.xlsx"
code_master = pd.read_excel(file_path)
code_master
```

```python
# 원본에 국적명 통합하기...merge()
pd.merge(sample_1, code_master, how="left")
```

```python
# 데이테프레임 변수명 : sample_2
# file : sample_2.xlsx

file_path = "./files/sample_2.xlsx"
sample_2 = pd.read_excel(file_path,
                         header = 1,
                         skipfooter = 2,
                         usecols = "A:C")
sample_2
```

```python
sample_2["기준년월"] = "2019-12"
sample_2
```

```python
sample_1.append(sample_2)
```

```python
sample_total = pd.concat([sample_1, sample_2], ignore_index=True) # ignore_index 순차적으로 학습
sample_total
```

```python
# 저장하기
# file : sample.xlsx

file_path = "./files/sample.xlsx"
sample_total.to_excel(file_path, index=False)
```
