### 딕셔너리 형태로 DataFrame으로 데이터 읽어들이기

```python
# 딕셔너리로 DataFrame으로 데이터 읽어들이기
# 1. tag의 정보를 가져오기
yad_list =soup.find_all("yadmnm")
sido_list =soup.find_all("sidocdnm")
addr_list =soup.find_all("addr")
telno_list =soup.find_all("telno")

#3. 전역변수 설정
yad_txt_list = []
sido_txt_list = []
addr_txt_list = []
telno_txt_list = []

# 2. for문 / 4. append로 변수 담기
for i in range(0, len(yad_list), 1) :
    #print(yad_list[i].text)
    yad_txt_list.append(yad_list[i].text)
    sido_txt_list.append(sido_list[i].text)
    addr_txt_list.append(addr_list[i].text)
    telno_txt_list.append(telno_list[i].text)

# print(yad_txt_list)
# print()
# print(sido_txt_list)
# print()
# print(addr_txt_list)
# print()
# print(telno_txt_list)

# 5. DataFrame에 담기(딕셔너리 형태 = {key(col) : value(값)})
df_temp = pd.DataFrame({"약국이름" : yad_txt_list,
                        "시도명" : sido_txt_list,
                        "주소" : addr_txt_list,
                        "전화번호" : telno_txt_list})
df_temp
```
