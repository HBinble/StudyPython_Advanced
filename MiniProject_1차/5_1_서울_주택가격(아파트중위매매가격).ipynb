### 서울 아파트 중위매매가격

```python
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
from matplotlib import font_manager, rc
font_path = "C:/Windows/Fonts/malgun.ttf"
font = font_manager.FontProperties(fname=font_path).get_name()
rc('font', family=font)
```

```python
file_path = "./data/주택가격/서울_중위매매가격_아파트.xlsx"
df = pd.read_excel(file_path, header=10)
df
```

```python
df1 = df.copy()
df1
```

```python
df2 = df1.fillna("서울")
```

```python
df3 = df2.rename(columns={"Unnamed: 1":"구"})
df3
```

```python
df4 = df3.T.reset_index()
df4
```

```python
df4["년도"] = df4["index"].str[0:5]
df4
```

```python
df5 = df4.drop(df4.columns[0], axis=1)
df5
```

```python
df5 = df5.rename(columns=df5.iloc[1])
df5
```

```python
df6 = df5.drop([0,1])
df6
```

```python
df6 = df6.rename(columns={"구": "년도"})
```

```python
df7 = df6.groupby("년도").mean()
df7
```

```python
df_ggd = df7.T
df_ggd
```

```python
df_ggd = df_ggd.astype(int)
```

```python
df_ggd.to_csv("./data/가공data/서울_주택가격(아파트중위매매가격).csv")
```

```python
plt.figure(figsize = (16,10))
sns.heatmap(df_ggd, annot=True, fmt = '.0f', cmap='GnBu')
plt.title('서울 아파트 중위 평균매매가격')
plt.show()
```

```python
df_ggd_sort = df_ggd.sort_values("2021년", ascending=False)
```

```python
plt.figure(figsize = (16,10))
sns.heatmap(df_ggd_sort, annot=True, fmt = '.0f', cmap='GnBu')
plt.title('서울 아파트 중위 평균매매가격')
plt.show()
```
