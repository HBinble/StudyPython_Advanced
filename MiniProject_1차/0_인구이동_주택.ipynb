```python
file_path = "./data/중위매매가격_종합.xlsx"
df = pd.read_excel(file_path, header=10)
df
```

```python
df1 = df.copy()
df1 = df1.drop(df1.columns[[1,2]], axis=1)
df1
```

```python
df_s_g = df1[(df1["지 역"] == "서울") | (df1["지 역"] == "경기")]
```

```python
df_s_g = df_s_g.reset_index(drop=True)
df_s_g.rename(columns={"지 역" : "지역"},inplace=True)
df_s_g
```

```python
df_s_g_T = df_s_g.T.reset_index()
df_s_g_T
```

```python
df_s_g_T["년도"] = df_s_g_T["index"].str[0:5]
df_s_g_T
```

```python
df_s_g_T_reindex = df_s_g_T.drop(df_s_g_T.columns[0], axis=1)
df_s_g_T_reindex.columns = ["서울", "경기", "년도"]
df_ggd = df_s_g_T_reindex.drop(index=0)
```

```python
df_ggd = df_ggd.groupby("년도").sum()
```

```python
df_ggd = df_ggd.drop("2022년")
```

```python
df_ggd.plot()
```
