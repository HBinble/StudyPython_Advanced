```python
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
from matplotlib import font_manager, rc
font_path = "C:/Windows/Fonts/malgun.ttf"
font = font_manager.FontProperties(fname=font_path).get_name()
rc('font', family=font)
```

```python
file_path = "./data/복지시설/경기도_노인여가복지시설(노인복지관)현황.csv"
df = pd.read_csv(file_path, encoding="EUC-KR")
df.info()
```

```python
df_sigu = df[["시군명", "이용회원수"]]
df_sigu
```

```python
df_sigu_group = df_sigu.groupby("시군명").sum()
df_sigu_group
```

```python
plt.rcParams['figure.figsize'] = (30,15)
plt.rcParams['lines.linewidth'] = 2
plt.rc("font", family = "Malgun Gothic", size=10)
df_sigu_group.plot(kind="barh")

plt.title("경기도_노인여가복지시설(노인복지관)현황", fontsize=20)
plt.xlabel("이용회원수", fontsize=20)
a = range(-1,32,1)

plt.xticks(fontsize = 20, rotation=0)
plt.yticks(fontsize = 20)

plt.legend(loc="upper right", fontsize=20)
plt.grid(True)

plt.show()
```

```python
df_sigu_group_asc = df_sigu_group.sort_values("이용회원수", ascending=False)
```

```python
df_sigu_group_top10 = df_sigu_group_asc.head(10)
df_sigu_group_top10
```

```python
plt.rcParams['figure.figsize'] = (30,15)
plt.rcParams['lines.linewidth'] = 2
plt.rc("font", family = "Malgun Gothic", size=10)
df_sigu_group_top10.plot(kind="bar")

plt.title("경기도_노인여가복지시설(노인복지관)현황", fontsize=20)
plt.xlabel("이용회원수", fontsize=20)
a = range(-1,32,1)

plt.xticks(fontsize = 20, rotation=0)
plt.yticks(fontsize = 20)

plt.legend(loc="upper right", fontsize=20)
plt.grid(True, alpha=0.3)

plt.show()
```
