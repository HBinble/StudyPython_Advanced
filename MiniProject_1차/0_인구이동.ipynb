### 주택보급률(아직 진행중)

```python
df.columns = ['지역', '주택보급률', '2010', '2011', '2012', '2013', '2014',
       '2015', '2016', '2017', '2018', '2019', '2020']
df
```

```python
seoul_jutek = df[3:6]
gyeonggi_jutek = df[27:30]
```

```python
seoul_jutek = seoul_jutek.fillna("서울")
seoul_jutek
```

```python
gyeonggi_jutek = gyeonggi_jutek.fillna("경기")
gyeonggi_jutek
```

```python
# 서울, 경기 주택보급률
jutek = pd.concat([seoul_jutek,gyeonggi_jutek], ignore_index=True)
jutek
```

```python
jutek = jutek.replace(',','',regex=True)
jutek
```

```python
jutek[['2010', '2011', '2012', '2013', '2014', '2015', '2016', '2017', '2018', '2019', '2020']] = \
    jutek[['2010', '2011', '2012', '2013', '2014', '2015', '2016', '2017', '2018', '2019', '2020']].astype(float)
```

```python
jutekT = jutek.groupby("주택보급률").get_group("주택수").T
jutekT.drop[0]
```
