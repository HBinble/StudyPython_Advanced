```python
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
file_path = "./data/주택가격/서울_아파트거래_연도별.xlsx"
df = pd.read_excel(file_path, header=10)
df
```

```python
df1 = df.copy()
df1
```

```python
df1 = df1.fillna("서울")
df1
```

```python
df1 = df1.rename(columns={"Unnamed: 1":"구"})
df1
```

```python
df2 = df1.drop(df1.columns[[2,3,4,5]],axis=1)
df2
```

```python
df3 = df2.drop(columns="지 역").reset_index(drop=True)
df3
```

```python
df3 = df3.drop(0)
```

```python
df4 = df3.set_index("구")
df4
```

```python
df5 = df4.sort_values("2021년", ascending=False)
```

```python
plt.figure(figsize = (16,10))
sns.heatmap(df5, annot=True, fmt = '.0f', cmap='rainbow')
plt.title('서울 아파트 거래량')
plt.show()
```
