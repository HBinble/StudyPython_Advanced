```python
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
from matplotlib import font_manager, rc
font_path = "C:/Windows/Fonts/malgun.ttf"
font = font_manager.FontProperties(fname=font_path).get_name()
rc('font', family=font)
```

```python
file_path = "./data/서울_경기_시군구별_이동자수.csv"
df = pd.read_csv(file_path, encoding="EUC_KR", skipfooter=13)
df
```

```python
df1 = df.copy()
df_gygi = df1.drop([0,1]).reset_index(drop=True)
df_gygi
```

```python
df_gygi.loc[27, "2010":"2012"] = 0
df_gygi
```

```python
df_gygi[['2010', '2011', '2012', '2013', '2014', '2015', '2016',
       '2017', '2018', '2019', '2020', '2021']] = df_gygi[['2010', '2011', '2012', '2013', '2014', '2015', '2016',
       '2017', '2018', '2019', '2020', '2021']].astype(float)
```

```python
df_gygi = df_gygi.set_index("행정구역(시군구)별")
```

```python
plt.figure(figsize = (16,10))
sns.heatmap(df_gygi, annot=True, fmt = '.0f', cmap='rocket_r')
plt.title('경기도별 인구이동 현황')
plt.show()
```

```python
df_ggu = df_gygi.sort_values("2021", ascending=False)
```

```python
plt.figure(figsize = (16,10))
sns.heatmap(df_ggu, annot=True, fmt = '.0f', cmap='rocket_r')
plt.title('경기도별 인구이동 현황')
plt.show()
```
