```python
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
from matplotlib import font_manager, rc
font_path = "C:/Windows/Fonts/malgun.ttf"
font = font_manager.FontProperties(fname=font_path).get_name()
rc('font', family=font)
```

```python
file_path = "./data/서울_경기_시군구별_이동자수.csv"
df = pd.read_csv(file_path, encoding="EUC_KR")
df
```

```python
df1 = df.copy()
df1
```

```python
df2 = df1.drop([0,1]).reset_index(drop=True)
df2
```

```python
df2.loc[27, "2010":"2012"] = 0
df2
```

```python
df2[['2010', '2011', '2012', '2013', '2014', '2015', '2016','2017', '2018', '2019', '2020', '2021']] = \
df2[['2010', '2011', '2012', '2013', '2014', '2015', '2016', '2017', '2018', '2019', '2020', '2021']].astype(int)
```

```python
df2.loc[41] = df2.iloc[27] + df2.iloc[29]
df2
```

```python
df2["행정구역(시군구)별"].replace({"여주시여주군":"여주시"}, inplace=True)
df2
```

```python
df3 = df2.replace(0, np.NaN)
df3
```

```python
df4 = df3.dropna(0)
df4
```

```python
df5 = df4.sort_values("행정구역(시군구)별").reset_index(drop=True)
df5
```

```python
df5.to_csv('./data/가공data/경기도전입인구(정제).csv', index=False)
```

```python
df_gygi_asc2021 = df5.sort_values(by="2021", ascending=False)
df_gygi_asc2021
```

```python
df_gygi_asc2021 = df_gygi_asc2021.set_index("행정구역(시군구)별")
```

```python
plt.figure(figsize = (16,10))
sns.heatmap(df_gygi_asc2021, annot=True, fmt = '.0f', cmap='rocket_r')
plt.title('2010~2021년 경기도 시별 전입인구 현황')
plt.show()
```

```python
df_df_gygi_asc2021_top8 = df_gygi_asc2021.head(8)
df_df_gygi_asc2021_top8
```

```python
plt.rcParams['figure.figsize'] = (30, 15)
plt.rcParams['lines.linewidth'] = 2
plt.rc("font", family = "Malgun Gothic", size=10)
df_df_gygi_asc2021_top8.T.plot()

plt.title("1")
plt.ylabel("1")

a = range(-1,32,1)
# plt.xticks([['2010', '2011', '2012', '2013', '2014', '2015', '2016','2017', '2018', '2019', '2020', '2021']])


plt.legend(loc="upper right")
plt.grid(True)

plt.show()
```
