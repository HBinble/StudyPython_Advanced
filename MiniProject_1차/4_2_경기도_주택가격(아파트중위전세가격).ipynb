### 경기도 아파트 중위매매가격

```python
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
from matplotlib import font_manager, rc
font_path = "C:/Windows/Fonts/malgun.ttf"
font = font_manager.FontProperties(fname=font_path).get_name()
rc('font', family=font)
```

```python
file_path = "./data/주택가격/경기도_중위전세가격_아파트.xlsx"
df = pd.read_excel(file_path, header=10)
df
```

```python
df1 = df.copy()
df1
```

```python
df2 = df1.drop(df1.columns[[0,2]],axis=1)
df3 = df2.dropna(0).reset_index(drop=True)
df3
```

```python
df4 = df3.rename(columns={"Unnamed: 1" : "경기도시군별"})
df4
```

```python
df5 = df4.T.reset_index()
df5
```

```python
df5["년도"] = df5["index"].str[0:5]
df5
```

```python
df6 = df5.drop(df5.columns[0], axis=1)
df6
```

```python
df6 = df6.rename(columns=df6.iloc[0])
```

```python
df6 = df6.drop(0)
```

```python
df6 = df6.rename(columns={"경기도시군": "년도"})
```

```python
df7 = df6.groupby("년도").mean()
df7
```

```python
df_ggd = df7.T
df_ggd
```

```python
df_ggd = df_ggd.astype(int)
```

```python
plt.figure(figsize = (16,10))
sns.heatmap(df_ggd, annot=True, fmt = '.0f', cmap='Pastel2')
plt.title('경기도 아파트 중위 평균전세가격')
plt.show()
```

```python
df_ggd_sort = df_ggd.sort_values("2021년", ascending=False)
```

```python
plt.figure(figsize = (16,10))
sns.heatmap(df_ggd_sort, annot=True, fmt = '.0f', cmap='Pastel2')
plt.title('경기도 아파트 중위 평균전세가격')
plt.show()
```
