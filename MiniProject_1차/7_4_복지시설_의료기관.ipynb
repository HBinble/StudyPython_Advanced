```python
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
from matplotlib import font_manager, rc
font_path = "C:/Windows/Fonts/malgun.ttf"
font = font_manager.FontProperties(fname=font_path).get_name()
rc('font', family=font)
```

```python
file_path = "./data/./복지시설/경기도 의료기관_2010~2021.csv"
df = pd.read_csv(file_path)
df
```

```python
df2 = df[["시군별(2)", "2010", "2011", "2012", "2013", "2014", "2015", "2016", "2017", "2018", "2019", "2020"]]
df2
```

```python
df3 = df2.drop(range(0,4,1)).reset_index(drop=True)
df3
```

```python
df3 = df3.replace({"-" : 0})
```

```python
df3[["2010", "2011", "2012", "2013", "2014", "2015", "2016", "2017", "2018", "2019", "2020"]] = \
    df3[["2010", "2011", "2012", "2013", "2014", "2015", "2016", "2017", "2018", "2019", "2020"]].astype(int)
```

```python
df3.loc[34] = df3.iloc[19] + df3.iloc[20]
df3
```

```python
df4 = df3.drop([19,20,22])
```

```python
df4 = df4.replace({"여주군여주시" : "여주시"})
```

```python
df5 = df4.rename(columns={"시군별(2)": "경기도시"})
```

```python
df6 = df5.reset_index(drop=True)
```

```python
df6_top10 = df6.sort_values("2020", ascending=False).head(10)
df6_top10
```

```python
df6_top10 = df6_top10.set_index("경기도시")
```

```python
df6_top10_181920 = df6_top10[["2018", "2019", "2020"]]
```

```python
df6_pivot = df6_top10.pivot_table(values="2020", index="경기도시")
```

```python
plt.rcParams['figure.figsize'] = (30,15)
plt.rcParams['lines.linewidth'] = 2
plt.rc("font", family = "Malgun Gothic", size=10)
df6_pivot.plot(kind="bar")

plt.title("경기도_의료기관 현황", fontsize=20)
plt.xlabel("의료기관수", fontsize=20)
a = range(-1,32,1)

plt.xticks(fontsize = 20, rotation=0)
plt.yticks(fontsize = 20)

plt.legend(loc="upper right", fontsize=20)
plt.grid(True)

plt.show()
```

```python
plt.rcParams['figure.figsize'] = (30,15)
plt.rcParams['lines.linewidth'] = 2
plt.rc("font", family = "Malgun Gothic", size=10)
df6_top10_181920.plot(kind="bar",stacked=True)

plt.title("경기도_의료기관 현황", fontsize=20)
plt.xlabel("의료기관수", fontsize=20)
a = range(-1,32,1)

plt.xticks(fontsize = 20, rotation=0)
plt.yticks(fontsize = 20)

plt.legend(loc="upper right", fontsize=20)
plt.grid(True, alpha=0.3)

plt.show()
```
