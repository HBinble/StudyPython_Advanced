```python
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
from matplotlib import font_manager, rc
font_path = "C:/Windows/Fonts/malgun.ttf"
font = font_manager.FontProperties(fname=font_path).get_name()
rc('font', family=font)
```

```python
file_path = "./data/주택가격/경기도_아파트거래_연도별.xlsx"
df = pd.read_excel(file_path, header=10)
df
```

```python
df1 = df.copy()
df1
```

```python
df2 = df1.drop(df1.columns[[0,2,3,4,5,6]],axis=1)
df3 = df2.dropna(0).reset_index(drop=True)
df3
```

```python
df4 = df3.rename(columns={"Unnamed: 1" : "경기도시군별"})
df4
```

```python
df5 = df4.replace({"-": 0})
df5
```

```python
df5.loc[32] = df5.iloc[27] + df5.iloc[28]
```

```python
df6 = df5.drop([27,28]).reset_index(drop=True)
```

```python
df7 = df6.set_index("경기도시군별")
df7
```

```python
df7.rename({"여주시여주군":"여주시"}, inplace=True)
df7
```

```python
df8 = df7.sort_values("2021년", ascending=False)
```

```python
plt.figure(figsize = (16,10))
sns.heatmap(df8, annot=True, fmt = '.0f', cmap='RdYlGn_r')
plt.title('경기도 아파트 거래량')
plt.show()
```

```python
df8_top15 = df8.head(15)
df8_top15
```

```python
plt.figure(figsize = (16,10))
sns.heatmap(df8_top15, annot=True, fmt = '.0f', cmap='RdYlGn_r')
plt.title('경기도 아파트 거래량 Top15')
plt.show()
```
