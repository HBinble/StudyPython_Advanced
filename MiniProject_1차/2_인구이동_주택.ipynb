```python
file_path = "./data/중위매매가격_종합.xlsx"
df = pd.read_excel(file_path, header=10)
df
```

```python
df1 = df.copy()
df1 = df1.drop(df1.columns[[1,2]], axis=1)
df1
```

```python
df_s_g = df1[(df1["지 역"] == "서울") | (df1["지 역"] == "경기")]
```

```python
df_s_g = df_s_g.reset_index(drop=True)
```

```python
df_s_g_T = df_s_g.set_index("지 역").T.reset_index()
df_s_g_T["년도"] = df_s_g_T["index"].str[0:5]
df_s_g_T = df_s_g_T.drop(df_s_g_T.columns[0], axis=1)
dfT = df_s_g_T.groupby("년도").sum()
df_s_g_price = dfT.drop("2022년")
df_s_g_price = df_s_g_price.reset_index()
```
