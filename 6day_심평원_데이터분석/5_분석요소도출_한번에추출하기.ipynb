```python
import pandas as pd
```

```python
file_path = "./files_new/df_20TABLE_kor_명세서일반내역.xlsx"
df_20TABLE_org = pd.read_excel(file_path)
df_20TABLE_org
df_month_age = df_20TABLE_org[["수진자연령","요양개시일자", "심결본인부담금"]]
df_month_age

df_age_any = pd.DataFrame()
for i in range(0, 101, 10):
    
    df_age = df_month_age[(df_month_age["수진자연령"] >= i) & \
                    (df_month_age["수진자연령"] < i+10)]
    df_age = df_age.reset_index(drop=True)
    if i < 10 :
        df_age["연령대"] = "9세이하"
    else : 
        df_age["연령대"] = "{}대".format(i)
    df_age
    df_age_any = pd.concat([df_age_any, df_age], ignore_index=True)

    df_age_any["요양개시일자"] = df_age_any["요양개시일자"].astype(str)
    df_age_any["요양개시월"] = df_age_any["요양개시일자"].str[4 : 6]

df_age_any
```

```python
file_path = "./files_new/df_20TABLE_kor_명세서일반내역.xlsx"
df_20TABLE_org = pd.read_excel(file_path)
df_month_age = df_20TABLE_org[["수진자연령","요양개시일자", "심결본인부담금"]]


df_age_any = pd.DataFrame()
for i in range(0, 101, 10):
    
    df_age = df_month_age[(df_month_age["수진자연령"] >= i) & \
                    (df_month_age["수진자연령"] < i+10)]
    df_age = df_age.reset_index(drop=True)
    df_age["연령대"] = "{}대이하".format(i+10)
    df_age
    df_age_any = pd.concat([df_age_any, df_age], ignore_index=True)

    df_age_any["요양개시일자"] = df_age_any["요양개시일자"].astype(str)
    df_age_any["요양개시월"] = df_age_any["요양개시일자"].str[4 : 6]

df_age_any
```

```python
file_path = "./files_new/df_20TABLE_kor_명세서일반내역.xlsx"
df_20TABLE_org = pd.read_excel(file_path)
df_month_age = df_20TABLE_org[["수진자연령","요양개시일자", "심결본인부담금"]]

df_list = {}
df_final = pd.DataFrame()

for i in range(0,91,10):
    df_list[f'{i}대'] = df_month_age[(df_month_age['수진자연령'] >= i+0) & (df_month_age['수진자연령'] <= i+9)].reset_index(drop =True)
    df_list[f'{i}대']['연령대'] = f'{i+9}세 이하'
    df_list[f'{i}대']['요양개시월'] = df_list[f'{i}대']['요양개시일자'].astype(str).str[4:6]
    df_final = pd.concat([df_final,df_list[f'{i}대']])

df_final
```
