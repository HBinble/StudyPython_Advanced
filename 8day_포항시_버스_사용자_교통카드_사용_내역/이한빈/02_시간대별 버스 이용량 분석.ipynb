## 기준월 및 기준일자별 버스 이용량 분석 비교

```python
pivot_month_day = df.pivot_table(values = '승객연령',
                             index = '승차일', columns='승차월',
                             aggfunc = 'count')
pivot_month_day = pivot_month_day.fillna(0)
pivot_month_day
```

```python
a = []
for i in range(0,32):
    b = i
    a.append(b)
    
print(a)
```

```python
plt.rcParams['figure.figsize'] = (20,10)

plt.rcParams['lines.linewidth'] = 2

plt.rc('font', family = 'Malgun Gothic', size = 10)

pivot_month_day.plot()

plt.title('기준월 및 기준일자별 버스 이용량')
plt.ylabel('버스 이용량')

plt.xticks(a)

plt.legend(loc='upper right')

plt.grid(True)

plt.show()
```

## 기준시간 및 시간(분)별 버스 이용량 분석 비교

### 10분 간격으로 분대를 만들어 비교

```python
df['승차분'].unique()
```

```python
# 승차분대
df['승차분대'] = 'm'

for i in range(0, 60, 10) : 
    if i == 0 :
        row_index = df.index[df['승차분'] < i+10]
        df.loc[row_index, '승차분대'] = '10분미만'

    if i >= 10 :
        row_index = df.index[(df['승차분'] >= i) & (df['승차분'] < i+10)]
        df.loc[row_index, '승차분대'] = '{}분대'.format(i)
        
df
```

```python
pivot_hour_min = df.pivot_table(values = '승객연령',
                             index = '승차시', columns='승차분대',
                             aggfunc = 'count')
pivot_hour_min = pivot_hour_min.fillna(0)
pivot_hour_min
```

```python
plt.rcParams['figure.figsize'] = (20,10)

plt.rcParams['lines.linewidth'] = 2

plt.rc('font', family = 'Malgun Gothic', size = 10)

pivot_hour_min.plot()

plt.title('기준시간 및 시간(분)별 버스 이용량')
plt.ylabel('버스 이용량')

plt.xticks(a)

plt.legend(loc='upper right')

plt.grid(True)

plt.show()
```

```python
plt.figure(figsize = (16,10))
sns.heatmap(pivot_hour_min, annot=True, fmt = '.0f', cmap='rocket_r')
plt.title('기준시간 및 시간(분)별 버스 이용량')
plt.show()
```

### 각 시간의 성격을 기준으로 시간대 만들어 비교

```python
df['승차시'].unique()
# 
```

```python
# 승차시간대
df['승차시간대'] = 'h'

row_index = df.index[(df['승차시'] >= 0) & (df['승차시'] <= 5)]
df.loc[row_index, '승차시간대'] = '새벽(0~5시)'

row_index = df.index[(df['승차시'] >= 6) & (df['승차시'] <= 8)]
df.loc[row_index, '승차시간대'] = '아침(6~8시)'

row_index = df.index[(df['승차시'] >= 9) & (df['승차시'] <= 12)]
df.loc[row_index, '승차시간대'] = '오전(9~12시)'

row_index = df.index[(df['승차시'] >= 13) & (df['승차시'] <= 16)]
df.loc[row_index, '승차시간대'] = '오후(13~16시)'

row_index = df.index[(df['승차시'] >= 17) & (df['승차시'] <= 19)]
df.loc[row_index, '승차시간대'] = '저녁(17~19시)'

row_index = df.index[(df['승차시'] >= 20) & (df['승차시'] <= 23)]
df.loc[row_index, '승차시간대'] = '밤(20~23시)'

df
```

```python
pivot_hour_min_g = df.pivot_table(values = '승객연령',
                             index = '승차시간대', columns='승차분대',
                             aggfunc = 'count')
pivot_hour_min_g = pivot_hour_min_g.fillna(0)
pivot_hour_min_g
```

```python
plt.rcParams['figure.figsize'] = (20,10)

plt.rcParams['lines.linewidth'] = 2

plt.rc('font', family = 'Malgun Gothic', size = 10)

pivot_hour_min_g.plot()

plt.title('기준시간 및 시간(분)별 버스 이용량')
plt.ylabel('버스 이용량')

# plt.xticks(a)

plt.legend(loc='upper right')

plt.grid(True)

plt.show()
```

```python
plt.figure(figsize = (16,10))
sns.heatmap(pivot_hour_min_g, annot=True, fmt = '.0f', cmap='rocket_r')
plt.title('기준시간 및 시간(분)별 버스 이용량')
plt.show()
```
