```python
# 극단치 값 계산하기
```

```python
import numpy as np
```

```python
# q1(25%), q3(75%) 데이터 확인하기(percentile)
q1, q3 = np.percentile(df_drop_allow["나이"], [25, 75])
print(q1)
print(q3)
```

```python
iqr = q3 - q1
iqr
```

```python
# 최소극단치 = q1 - 1.5*iqr / 최대극단치 = q3 + 1.5*iqr
```

```python
#최대극단치 변수 : upper_bound
upper_bound = q3 + (iqr * 1.5)
upper_bound
```

```python
#최소극단치 변수 : lower_bound
lower_bound = q1 - (iqr * 1.5)
lower_bound
```

```python
# 나이 데이터 중에 최대/최소 극단치를 벗어나는 데이터 조회하기
condition = (df_drop_allow["나이"] >= upper_bound) | (df_drop_allow["나이"] <= lower_bound)
print(len(df_drop_allow[condition]))
df_drop_allow[condition]
```

```python
# 이상치 함수 만들기

import numpy as np

def outliers_iqr(dframe, data) :
    # q1(25%), q3(75%) 데이터 확인하기
    q1, q3 = np.percentile(data, [25, 75])
    iqr = q3 - q1
    upper_bound = q3 + (iqr * 1.5)
    lower_bound = q1 - (iqr * 1.5)

    condition = ((data > upper_bound) | \
                        (data < lower_bound))

    print(upper_bound)
    print(lower_bound)
    print(len(dframe[condition]))
    return dframe[condition]

```

```python
temp = outliers_iqr(df_drop_allow, df_drop_allow["이용거리"] )
temp
```
