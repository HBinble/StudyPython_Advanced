```python
# 기준년월 컬럼을 만들고 값은 2019-01로 넣어주세요
kto_201901["기준년월"] = "2019-01"
kto_201901
```

```python
# 특정 컬럼의 고유한 데이터만 조회하기
kto_201901["국적"].unique()
```

```python
continents_list = ['아시아주', '미주', '구주', '대양주',
                    '아프리카주', '기타대륙', '교포소계']
continents_list
```

```python
# 국적 데이터에서 continents_list에 대륙 이름이 없는 것만 필터링하기..
condition = kto_201901["국적"].isin(continents_list)
condition
```

```python
kto_201901_country = kto_201901[condition == False]
kto_201901_country.head(1)
```

```python
kto_201901_country["국적"].unique()
```

```python
kto_201901_country_newindex = kto_201901_country.reset_index(drop=True) # drop = True 기존의 index 값 삭제
kto_201901_country_newindex
```

```python
# 아시아, 아메리카, 유럽, 오세아니아, 아프리카, 기타대륙, 교포
continents = ["아시아"] * 25 + ["아메리카"] * 5 + ["유럽"] * 23 + \
             ["오세아니아"] * 3 + ["아프리카"] * 2 + ["기타대륙"] + ["교포"]
print(len(continents))
continents
```

```python
kto_201901_country_newindex["대륙"] = continents
kto_201901_country_newindex
```

```python
kto_201901_country_newindex.head()
```

```python
# 각 국가별 국내에 관광을 위해 입국하는 관광객 비율(%)
# 관광객비율(%) = 관광/계 *100 (소숫점 첫째자리까지)
kto_201901_country_newindex["관광객비율(%)"] = round(kto_201901_country_newindex["관광"] / kto_201901_country_newindex["계"] *100, 1)
kto_201901_country_newindex.sort_values(by = "관광객비율(%)", ascending=False)
```

```python
# 대륙별 관광객비율(%)의 평균 (pivot_table)
kto_201901_country_newindex.pivot_table(values = "관광객비율(%)",
                                                index = "대륙",
                                                aggfunc = "mean")
```

```python
kto_201901_country_newindex[kto_201901_country_newindex["국적"] == "중국"]
```

```python
kto_201901_country_newindex.head()
```

```python
# 컬럼명 : 전체비율(%) : 전체 관광 대비 각 국가별 관광
kto_201901_country_newindex["전체비율(%)"] = \
                            round(kto_201901_country_newindex["관광"] / \
                                  sum(kto_201901_country_newindex["관광"]) * 100, 1)
kto_201901_country_newindex.head()
```

### 문제 풀이

```python
df = pd.DataFrame()

for y in range(2010, 2020, 1) : 
      for m in range(1, 13, 1) :
            try : 
                  # 처리필요
                  file_nm = str(y) + str(m).zfill(2)
                  file_path = "./files/kto_{}.xlsx".format(file_nm)
                  kto_201901 = pd.read_excel(file_path, header=1, skipfooter=4, usecols="A:G")

                  #처리필요
                  kto_201901["기준년월"] = str(y) + "-" + str(m).zfill(2)

                  continents_list = ['아시아주', '미주', '구주', '대양주',
                                    '아프리카주', '기타대륙', '교포소계']

                  condition = kto_201901["국적"].isin(continents_list)
                  kto_201901_country = kto_201901[condition == False]
                  kto_201901_country_newindex = kto_201901_country.reset_index(drop=True)
                  continents = ["아시아"] * 25 + ["아메리카"] * 5 + ["유럽"] * 23 + \
                              ["오세아니아"] * 3 + ["아프리카"] * 2 + ["기타대륙"] + ["교포"]
                  kto_201901_country_newindex["대륙"] = continents
                  kto_201901_country_newindex["관광객비율(%)"] = round(kto_201901_country_newindex["관광"] / kto_201901_country_newindex["계"] *100, 1)
                  kto_201901_country_newindex["전체비율(%)"] = \
                                          round(kto_201901_country_newindex["관광"] / \
                                                sum(kto_201901_country_newindex["관광"]) * 100, 1)
                  df = pd.concat([df, kto_201901_country_newindex], ignore_index=True)

            except :
                  pass
```
