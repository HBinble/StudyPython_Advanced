```python
code_master = pd.read_excel("./files2/sample_codemaster.xlsx")
i = 0
sample_total = pd.DataFrame()

for y in range(2019, 2021, 1):
    for m in range(1, 13, 1):
        # print(str(y) + "-" + str(m))

        file_path = "./files2/sample({}).xlsx".format(i)
        sample = pd.read_excel(file_path, header=1, skipfooter=2, usecols="A:C")
        i += 1

        sample["기준년월"] = str(y) + "-" + str(m).zfill(2)
        sample_end = pd.merge(sample, code_master, how="left")

        sample_total = pd.concat([sample_total, sample_end])

sample_total
```

```python
df = pd.DataFrame()

for y in range(2010, 2020, 1) :
    for m in range(1, 13, 1) :

        file_nm = str(y) + str(m).zfill(2)
        file_path = "./files/kto_{}.xlsx".format(file_nm)
        kto_201001 = pd.read_excel(file_path, header=1, skipfooter=4, usecols="A:G")
        
        kto_201001["기준년월"] = str(y) + "-" + str(m).zfill(2)

        continents_list = ['아시아주', '미주', '구주', '대양주',
                            '아프리카주', '기타대륙', '교포소계']
        condition = kto_201001["국적"].isin(continents_list)                    
        kto_201001_country = kto_201001[condition == False]
        kto_201001_country_newindex = kto_201001_country.reset_index(drop=True)
        continents = ["아시아"] * 25 + ["아메리카"] * 5 + ["유럽"] * 23 + \
                    ["오세아니아"] * 3 + ["아프리카"] * 2 + ["기타대륙"] + ["교포"]
        kto_201001_country_newindex["대륙"] = continents
        kto_201001_country_newindex["관광객비율(%)"] = round(kto_201001_country_newindex["관광"] / kto_201001_country_newindex["계"] * 100 ,1)
        kto_201001_country_newindex["전체비율(%)"] = \
            round(kto_201001_country_newindex["관광"] / \
                sum(kto_201001_country_newindex["관광"]) * 100,1) 
df
```
